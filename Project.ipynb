```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Placement_Data_Full_Class.csv')
df
```

```python
df.fillna(0,inplace=True)
```

```python
df.duplicated().sum()
```

```python
#drop id ,gender
df.drop(['sl_no','gender'],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['ssc_b']=le.fit_transform(df['ssc_b'])
df['hsc_b']=le.fit_transform(df['hsc_b'])
df['hsc_s']=le.fit_transform(df['hsc_s'])
df['degree_t']=le.fit_transform(df['degree_t'])
df['workex']=le.fit_transform(df['workex'])
df['specialisation']=le.fit_transform(df['specialisation'])
df['status']=le.fit_transform(df['status'])
df
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.scatterplot(x='degree_p', y='etest_p', data=df)
```

```python
df.drop(['ssc_b','hsc_b'],axis=1,inplace=True)
```

```python
df
```

```python
a=df.drop('salary',axis=1,inplace=True)
```

```python
X=df.drop('status',axis=1)
y=df['status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=33)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
random=RandomForestClassifier()
random.fit(X_train,y_train)
random_pred=random.predict(X_test)

print('Accurancy =',accuracy_score(y_test,random_pred))
print(classification_report(y_test,random_pred))
```

```python
#svc
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,y_train)
svc_pred=svc.predict(X_test)

print('Accurancy =',accuracy_score(y_test,svc_pred))
print(classification_report(y_test,svc_pred))
```

```python
#decisition tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
dt_pred=dt.predict(X_test)

print('Accurancy =',accuracy_score(y_test,dt_pred))
print(classification_report(y_test,dt_pred))
```

```python
#logistic regreesion
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
lr_pred=lr.predict(X_test)

print('Accurancy =',accuracy_score(y_test,lr_pred))
print(classification_report(y_test,lr_pred))
```

```python

import joblib
joblib.dump(lr, 'placement_prediction_model.pkl')
model = joblib.load('placement_prediction_model.pkl')
```
