```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
df
```

```python
#split and testing y(out),training X(in)
X=df.drop('Species',axis=1)
y=df['Species']
```

```python
#train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=33)
```

```python
from sklearn.metrics import accuracy_score,classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
knn_grade=knn.predict(X_test)



print('Accurancy= ',accuracy_score(y_test,knn_grade))
target_names=['setosa','versicolor','virginica']
print(classification_report(y_test,knn_grade,target_names=target_names))
```

```python
#SVc
svc=SVC()
svc.fit(X_train,y_train)
svc_grade=svc.predict(X_test)
print('Accurancy= ',accuracy_score(y_test,svc_grade))

target_names=['setosa','versicolor','virginica']
print(classification_report(y_test,svc_grade,target_names=target_names))
```

```python
#decision tree
tree=DecisionTreeClassifier()
tree.fit(X_train,y_train)
tree_grade=tree.predict(X_test)
print('Accurancy= ',accuracy_score(y_test,tree_grade))
print(classification_report(y_test,tree_grade,target_names=target_names))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
log=LogisticRegression()#if optimal solution not get in the
log.fit(X_train,y_train)
log_grade=log.predict(X_test)
print('Accurancy= ',accuracy_score(y_test,log_grade))

print(classification_report(y_test,log_grade,target_names=target_names))
```

```python
rand=RandomForestClassifier(120)
rand.fit(X_train,y_train)
rand_grade=rand.predict(X_test)
print('Accurancy= ',accuracy_score(y_test,rand_grade))

print(classification_report(y_test,rand_grade,target_names=target_names))
```
