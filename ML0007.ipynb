```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Mall_Customers.csv')
df
```

```python
from sklearn.cluster import KMeans
```

```python
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',data=df)
```

```python
# Select multiple columns by passing a list of column names
X=df[['Annual Income (k$)','Spending Score (1-100)']]

kmeans=KMeans(n_clusters=5,random_state=42)
df['Cluster']=kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(10,6))
plt.figure(figsize=(10,6))
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',hue='Cluster',data=df,palette='rainbow')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering')
plt.show()
```

```python
sns.scatterplot(x='Age',y='Spending Score (1-100)',data=df)
```

```python
X=df[['Age','Spending Score (1-100)']]

kmeans=KMeans(n_clusters=4,random_state=42)
df['Cluster']=kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(10,6))
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age',y='Spending Score (1-100)',hue='Cluster',data=df,palette='rainbow')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering')
plt.show()
```

```python
#elbow graph
sse=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=42)
  kmeans.fit(X)
  sse.append(kmeans.inertia_)


plt.plot(range(1,11),sse,marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
